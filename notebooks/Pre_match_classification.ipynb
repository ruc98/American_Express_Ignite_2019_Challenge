```python
##########################################################################
######## ON KAGGLE ############
##########################################################################
import numpy as np
import pandas as pd
import glob
import os

from random import shuffle
```

```python
# load and process data
path = r'odi_csv/'
all_files = glob.glob(path + "/*.csv")
shuffle(all_files)
target=[]

team1=[]
team2=[]
date=[]
series=[]
venue=[]
city=[]
toss_winner=[]
toss_decision=[]
count=0
for filename in all_files:
    df = pd.read_csv(filename,usecols=[0])
    skips = df.loc[ 'info' , : ].shape[0]

    df = pd.read_csv(filename,nrows=skips,skiprows=1,header=None)
    df = df.drop(columns=0).set_index(df.columns[1])
    df0=df
    winteam=None
    if 'winner' in df.index:
        winteam = df.loc['winner',:].values[0]

    attr1=None
    attr2=None
    coln='team'
    if coln in df.index:
        attr1 = df.loc[coln,:].values[0][0]
        attr2 = df.loc[coln,:].values[1][0]
    team1.append(attr1)
    team2.append(attr2)



    attr=None
    coln='date'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]
    date.append(attr)
    attr=None
    coln='series'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]    
    series.append(attr)
    attr=None
    coln='venue'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]    
    venue.append(attr)
    attr=None
    coln='city'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]    
    city.append(attr)
    attr=None
    coln='toss_winner'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]    
    toss_winner.append(attr)
    attr=None
    coln='toss_decision'
    if coln in df.index:
        if df.loc[coln,:].shape[0] ==1:
            attr = df.loc[coln,:].values[0]
        if df.loc[coln,:].shape[0] ==2:
            attr = df.loc[coln,:].values[0][0]
    toss_decision.append(attr)

    
    df = pd.read_csv(filename,skiprows=skips+1,header=None)
    
    df2 = df[df.columns[[1,2,7,8]]].set_index(df.columns[1]).drop(index=1)
    if df2.shape[0]>0:
        i2team = df[df.columns[[1,3]]].set_index(df.columns[1]).drop(index=1).values[0,0]
        if (i2team==winteam):
            target.append(1)
        else:
            target.append(0)
    else:
        target.append(0)
    count+=1
```

```python
df = pd.DataFrame(list(zip(team1,team2,date,series, venue, city, toss_winner, toss_decision,target)), columns =['team1','team2','date','series','venue', 'city', 'toss_winner', 'toss_decision','target']) 
dropteams=['Africa XI','Asia XI','Oman','United States of America']
for dropteam in dropteams:
    df.drop(df.loc[df['team1']==dropteam].index, inplace=True)
    df.drop(df.loc[df['team2']==dropteam].index, inplace=True)
```

```python
df=pd.concat([df,teamsdummies],axis=1)
df
```

```python
citydummies=pd.get_dummies(df['city'])
df=pd.concat([df,citydummies],axis=1)
df
```

```python
tossdummies=pd.get_dummies(df['toss_decision'])
df=pd.concat([df,tossdummies],axis=1)
df
```

```python
intdates=pd.to_timedelta(pd.to_datetime(df['date']))
intdates-=intdates.min()
intdates=(intdates.astype(int)/10**9).astype(int)
df.drop(['date'], axis=1,inplace=True)
df=pd.concat([df,intdates],axis=1)
```

```python
df.drop(['team1','team2','series','venue','city','toss_winner','toss_decision'], axis=1,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

X = df.drop(columns=['target']).values
y = df['target'].values
# split data into train and test sets
seed = 10
test_size = 0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
model = AdaBoostClassifier()
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_pred = model.predict(X_test)
trainacc = accuracy_score(y_train,y_train_pred)
testacc  = accuracy_score(y_test, y_pred)
print("Train Accuracy: %.2f%%" % (trainacc * 100.0))
print("Test Accuracy: %.2f%%" % (testacc * 100.0))
```
